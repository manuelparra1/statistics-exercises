````python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import viz # curriculum viz example code

np.random.seed(123)
````

````python
die_distribution = stats.randint(1,7)
````

````python
die_distribution.rvs()
````

````python
die_distribution.rvs(5)
````

````python
die_distribution.rvs((5,5))
````

````python
n = 10_000
x = die_distribution.rvs(n)
````

````python
plt.hist(x, bins=range(0,9), align='left',width=.95, edgecolor='black')
plt.title(f'Outcome of {n:,} Dice Rolls')
````

````python
# USE a PMF to find the probality that a roll = 3
# SINGULAR PROBABILITY

die_distribution.pmf(3)
````

# Cumulative Density Function

````python
#   C.  D.   F.   
````

````python
# cumulative density function
# Use a CDF for the problity ata roll is 3 or lower
# INCLUSIVE
# EVERYTHING TO THE LEFT
# LESS THAN
die_distribution.cdf(3)
````

````python
viz.distributions_example1(die_distribution)
````

# Percent Point Function

````python
#.   P.   P.   F.  
## Use a PPF function to fuind the value asscociated with 5/6
##    *exclusive*
# INCLUSIVE
#     REVERSE of CDF
# EVERYTHING TO THE LEFT
# LESS THAN
die_distribution.ppf(5/6)
````

# Survival Function

````python
# Survival Function
#
# outputs = probabilty greater than (above, exclusive)value
# input = value

die_distribution.sf(4)
````

````python
# Inverse Survival Function
die_distribution.isf(1/3)
````
