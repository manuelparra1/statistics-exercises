# ANOVA

Comparing more than 2 groups

LEVENE

Test for variance A,B,C

```python
df=data('cake')
```

```python
a_temp = df[df['recipe']=='A']['temp']
b_temp = df[df['recipe']=='B']['temp']
c_temp = df[df['recipe']=='C']['temp']
```

```python
a_temp.hist()
plt.show()
```

```python
b_temp.hist()
plt.show()
```

```python
c_temp.hist()
plt.show()
```

```python
print(a_temp.var())
print(b_temp.var())
print(c_temp.var())
```

```python
stats.levene(a_temp,b_temp,c_temp)
```

```python
stats.f_oneway(a_temp,b_temp,c_temp)
```

```python
alpha = .05
f,p=stats.f_oneway(a_temp,b_temp,c_temp)
p<alpha
```
